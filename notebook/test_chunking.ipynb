```python
import sys
import os
import re
sys.path.append(os.path.abspath(".."))
from chunker import PDFChunker
```

```python
chunker = PDFChunker({'chunk_size': 300,
                      'chunk_overlap': 50,
                      'min_title_font_size': 14,
                      'min_chunk_len': 200})
```

```python
pages = chunker.extract_text_from_pdf("../Document.pdf")
page = pages[15]["text"]
page
```

```python
sentences = re.split(r'(?<=[\.\!\?])\s+', page)

for i, sentence in enumerate(sentences):
    print("Sentence " + str(i) + " : " + sentence + "\n")
```
